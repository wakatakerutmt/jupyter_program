```python
# 必要なimport
import cv2
import numpy as np
```

```python
# 画像の読み込み　imread()
img = cv2.imread("assets/imori.jpg")
```

```python
img.shape
# 画像のサイズ情報をとるにはimg.shapeを使います。 
# これは(高さ、横、チャネル)の順になっていて、つまり 縦128ピクセル、横128ピクセル、3チャネル(青、緑、赤)を意味します。
```

```python
img.dtype
# 型を取る　 uint8とは符号なしintで8ビット
# 画像はRGB成分がそれぞれ通常0から255の256段階で表現される 
# 例えば、赤のピクセルは(R,G,B) = (255, 0, 0)、白は(R,G,B) = (255, 255, 255)
# 画像にする時は必ずこの型にしないといけない
```

```python
# 画像をいじる時は、

# 1 cv2.imreadで読み込む
# 2 型を一度np.float32など小数点型に変換する
# 3 画像をいじる
# 4 画素が0未満のものは0, 255を超えたものは255に直す。（超重要）
# 5 型をnp.uint8に変換して表示・保存する

# という手順を踏む
```

```python
# 画像のいじり方

img[20, 30]

# 例えば、x=30、y=20の画素値をいじりたい時は、こうする 
# 画素値はBGRの順に並んでいる
# array()とはnumpy形式になっていることを意味する。つまり、OpenCVはnumpyのラッパーである
```

```python
# さらに、x=30、y=20、のG成分をとる時はこうする
img[20, 30, 1]  # 3つ目が 0:B 1:G 2:R  でRGBの逆順になっていることに注意
```

```python
# 画像のいじり実践
# 画像の左上(x=[0, 50], y = [0, 50])を黒にする

img2 = img.copy() # copy()でなく img2 = img とするとアドレスが保存されimgにも影響が出る（Numpyのスライスと同様である）
img2[:50, :50] = 0
imshow_origin(img2)
```

```python

# 次に先程の　4. 画素が0未満のものは0, 255を超えたものは255に直す　に関して実験
# 例えば、画像を一度 float32型にして、一部分のB成分を260という値に変えるとする

# ↓のコメントを外す
# img2 = img.copy().astype(np.float32)
# img2[60:100, 60:100, 0] = 260
# img2 = img2.copy().astype(np.uint8)
# imshow_origin(img2)

# uint8型は[0, 255] の範囲しか取り得ないので、これをuint8型に直すとこんな風になる。 イモリの顔の一部分が黄色くなってしまう。
# これは、260をuint8型に直すと260 - 256 が起きて、B=4となってしまうため 
# これが原因で画素値がおかしくなることが多々起きてしまう。 なので、4の操作が必要。

```

```python
# 保存には cv2.imwrite()を使いましょう。

cv2.imwrite("sample.jpg", img2)

# 例えば先程のimg2を sample.jpg という名前で保存する時はこうしましょう。
# Trueとなれば保存に成功で、これで同じフォルダ内にsample.jpgというファイルができています。
```

```python
# 練習
# 画像の左半分上のRとBを入れ替えて表示する

import cv2
img = cv2.imread('assets/imori.jpg')
img2 = img.copy()
H, W, C = img2.shape
img2[:H//2, :W//2] = img2[:H//2, :W//2, (2,1,0)]
imshow_origin(img2)
```
