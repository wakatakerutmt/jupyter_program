```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(25)
```

```python
dataset
```

```python
# rugplotを書きます
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
# ヒストグラムを重ねます。
plt.hist(dataset,alpha=0.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset);

# X軸を競ってい
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 全体を等間隔に100分割します。
x_axis = np.linspace(x_min,x_max,100)

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2

kernel_list = []

for data_point in dataset:
    
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)

    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

```python
# 手動でKDEを作ります。
sum_of_kde = np.sum(kernel_list,axis=0)

fig = plt.plot(x_axis,sum_of_kde,color='indianred')

sns.rugplot(dataset,c = 'indianred')

plt.yticks([])

plt.suptitle("Sum of the Basis Functions")
```

```python

# これが1行で出来ます。
sns.kdeplot(dataset)
```

```python

sns.rugplot(dataset,color='black')
# バンド幅を変えてみましょう。
for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset,bw=bw,label=bw)
```

```python

# カーネル関数の説明（英語）
# http://en.wikipedia.org/wiki/Kernel_(statistics)
```

```python
kernel_options = ["biw", "cos", "epa", "gau", "tri", "triw"]

for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern)
```

```python

for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern,shade=True,alpha=0.5)
```

```python

# 軸を入れ替えることもできます。
sns.kdeplot(dataset,vertical=True)
```

```python

# cumulative distribution function (CDF)
# 累積分布関数
# http://en.wikipedia.org/wiki/Cumulative_distribution_function
    
plt.hist(dataset, cumulative=True)
```

```python
sns.kdeplot(dataset,cumulative=True)
```

```python
# 2次元平面上の中心
mean = [0,0]
# それぞれの分散を決めます。
cov = [[1,0],[0,100]]
# これに従う多変量正規分布
dataset2 = np.random.multivariate_normal(mean,cov,1000)
# DataFrameにしておきましょう。
dframe = pd.DataFrame(dataset2,columns=['X','Y'])
# プロットします。SeabornとPandasの相性は抜群
sns.kdeplot(dframe)
```

```python
# 軸ごとにデータをわたせます。
sns.kdeplot(dframe.X,dframe.Y)
```

```python
# 軸ごとにデータをわたせます。
sns.kdeplot(dframe.X,dframe.Y, shade=True)
```

```python
# バンド幅を変えられます。
sns.kdeplot(dframe,bw=1)
```

```python

# 文字列でも渡せます。
sns.kdeplot(dframe,bw='silverman')
```

```python
# 同時分布の推定も可能です。
sns.jointplot('X','Y',dframe,kind='kde')
```

```python

# Violin/Viola plot はbox plotのシンプルさとKDEの情報量を兼ね備えます。
```

```python
# 正規分布に従う乱数です。
data1 = stats.norm(0,5).rvs(100)
```

```python
# γ分布に従う乱数を生成します。
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# ヴァイオリンプロットを描くと違いが分かります。
sns.violinplot(data =[data1,data2])
```

```python
# バンド幅を細かくしてみましょう。
sns.violinplot(data=data2,bw=0.01)
```

```python
# rugプロットと同じような機能もあります。
sns.violinplot(data=data1,inner="stick")
```

```python
# ひさしぶりにかく
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Seabornにサンプルデータがあります。
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# lmplot() ｘ=total bill y=tipsでプロットします。
sns.lmplot("total_bill","tip",tips)
# liner modelの略？回帰直線が書かれる
```

```python

# グラフごとにパラメータを変えられます。
sns.lmplot("total_bill","tip",tips,
                   scatter_kws={'marker':'o', 'color':'indianred'},
                   line_kws={'linewidth':1, 'color':'blue'})
```

```python
# 4次関数で回帰曲線をひくこともできます。
sns.lmplot("total_bill", "tip", tips, order=4,
           scatter_kws={"marker": "o", "color": "indianred"},
           line_kws={"linewidth": 1, "color": "blue"})
```

```python

# 単なるプロットもできます。
sns.lmplot("total_bill", "tip", tips,fit_reg=False)
```

```python
# 離散的な値でもlmplot()は使えます。

# チップの割合を計算します。
tips["tip_pect"]=100*(tips['tip']/tips['total_bill'])
tips.head()
```

```python
sns.lmplot("size", "tip_pect", tips)
```

```python
sns.lmplot("size", "tip_pect", tips, x_jitter=0.2)
```

```python
# 特定の推定量を描画することもできます。
sns.lmplot("size", "tip_pect", tips, x_estimator=np.mean);
```

```python
# hueが便利です。
sns.lmplot("total_bill", "tip_pect", tips, hue="sex",markers=["x","o"])
```

```python
sns.lmplot("total_bill", "tip_pect", tips, hue="day")
```

```python
# LOESSの説明（英語）です。
url = 'http://en.wikipedia.org/wiki/Local_regression'
sns.lmplot("total_bill", "tip_pect", tips, lowess=True, line_kws={"color": 'black'});
```

```python
# lmplot() は実は、もっと低レベルな関数regplotを使っています。
sns.regplot("total_bill","tip_pect",tips)
```

```python
# 描画のエリアを分割することもできます。
fig, (axis1,axis2) = plt.subplots(1,2,sharey =True)

sns.regplot("total_bill","tip_pect",tips,ax=axis1)
sns.violinplot(y='tip_pect', x='size', data=tips.sort('size') ,ax=axis2)
```

```python
flight_dframe = sns.load_dataset('flights')
```

```python
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot('month','year','passengers')
```

```python
flight_dframe
```

```python
sns.heatmap(flight_dframe)
```

```python

# 数字を書き込めます。
sns.heatmap(flight_dframe, annot=True, fmt='d')
# annot アノテーションのこと
```

```python

#中心を指定して、色を変えられます。
sns.heatmap(flight_dframe, center=flight_dframe.loc['January',1955])
```

```python

f, (axis1, axis2) = plt.subplots(2,1)

yearly_flights = flight_dframe.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_dframe = pd.concat((years, flights), axis=1)
year_dframe.columns = ['Year', 'Flights']

sns.barplot('Year', y='Flights', data=year_dframe, ax=axis1)
sns.heatmap(flight_dframe, cmap='Blues', ax=axis2, cbar_kws={'orientation':'horizontal'})
```
